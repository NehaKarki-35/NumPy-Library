# Reshaping in array

***arr.shape***

***Using arr.reshape() will give a new shape to an array without changing the data. Just remember that when you use the reshape method, the array you want to produce needs to have the same number of elements as the original array. If you start with an array with 12 elements, you’ll need to make sure that your new array also has a total of 12 elements.***

```python
import numpy as np
```

```python
a=np.arange(6)
a
```

```python
b=a.reshape(3,2)
b
```

```python
c=b.reshape(4,1)
```

```python
c=b.reshape(6,1)
c
```

```python
a=a.reshape(1,6)
a
```

```python
a=a.reshape(6,1)
a
```

# Brodcasting

***Broadcasting is a mechanism that allows NumPy to perform operations on arrays of different shapes***

***There are many broadcasting rules, which you can learn from the numpy documentation***

```python
distance_in_km=np.array([2,4,6,8])
single_value=np.array(1000)
distance_in_km *  single_value
```

```python
print(single_value.shape)
```

```python
print(single_value.size)
```

```python
print(distance_in_km.shape)
```

```python
print(distance_in_km.size)
```

```python
print(distance_in_km.ndim)
```

```python
print(single_value.ndim)
```
